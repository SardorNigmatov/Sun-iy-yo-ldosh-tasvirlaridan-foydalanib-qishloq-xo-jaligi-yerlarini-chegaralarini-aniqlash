```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\BMIProject\files_Adam\train_log.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Train va Validation Loss grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainLoss'], label='Train Loss')
plt.plot(df['Epoch'], df['ValLoss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train va Validation Loss grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train va Validation F1-Score grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainF1'], label='Train F1-Score')
plt.plot(df['Epoch'], df['ValF1'], label='Validation F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train va Validation F1-Score grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train va Validation Jaccard Index grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainJaccard'], label='Train Jaccard Index')
plt.plot(df['Epoch'], df['ValJaccard'], label='Validation Jaccard Index')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train va Validation Jaccard Index grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train va Validation Precision grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainPrecision'], label='Train Precision')
plt.plot(df['Epoch'], df['ValPrecision'], label='Validation Precision')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train va Validation Precision grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train va Validation Recall grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainRecall'], label='Train Recall')
plt.plot(df['Epoch'], df['ValRecall'], label='Validation Recall')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train va Validation Recall grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Train Accurcay grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['TrainRecall'], label='Train Recall')
plt.plot(df['Epoch'], df['TrainPrecision'], label='Train Precision')
plt.plot(df['Epoch'], df['TrainF1'], label='Train F1-Score')
plt.plot(df['Epoch'], df['TrainJaccard'], label='Train Jaccard Index')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train accurcay grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Val Accurcay grafigini chizish
plt.figure(figsize=(10, 6))
plt.plot(df['Epoch'], df['ValRecall'], label='Val Recall')
plt.plot(df['Epoch'], df['ValPrecision'], label='Val Precision')
plt.plot(df['Epoch'], df['ValF1'], label='Val F1-Score')
plt.plot(df['Epoch'], df['ValJaccard'], label='Val Jaccard Index')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation accurcay grafigi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['TrainRecall'], df['TrainPrecision'], label='Train Precision-Recall', marker='o')
plt.plot(df['ValRecall'], df['ValPrecision'], label='Validation Precision-Recall', marker='s')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision va Recall bogliqligi grafigi')
plt.legend()
plt.grid(True)
plt.show()
```
